```python
tick = "TSLA"
```

```python
print(f"Selected ticker: {tick}")
```

```python
history = yf.Ticker(tick).history(period="3mo")
```

```python
print(f"3 Month low {round(history.Low.min())}")
print(f"3 Month low {round(history.High.min())}")
```
